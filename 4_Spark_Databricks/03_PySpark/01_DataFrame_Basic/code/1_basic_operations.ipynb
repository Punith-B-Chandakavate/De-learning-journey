```python
spark
from pyspark.sql.functions import col
```

```python
df = spark.table('workspace.default.movies')
df.show(5)
```

```python
display(df)
```

```python
colum = df.columns
display(colum)
```

```python
df.count()
```

```python
display(df.summary())
```

```python
df_trime = df.select('title', "industry")
df_trime.show(3, truncate=True)
```

```python
df_filtered = df.filter((df.release_year >= 2000) & (df.release_year <= 2005))
display(df_filtered)
```

```python
unique_industry = df.select('industry').distinct()
display(unique_industry)
```

```python
df_with_profit = df.withColumn('profit', col('revenue') - col('budget'))
display(df_with_profit)
```

```python
df = df.withColumnRenamed('revenue', 'total_revenue')
df.printSchema()
```
